```python
import numpy as np
url = "https://raw.githubusercontent.com/allanspadini/numpy/dados/citrus.csv"
dado = np.loadtxt(url, delimiter=",", skiprows=1, usecols=np.arange(1, 6, 1))
dado
```

```python
diametro_laranja = dado[:5000, 0]
diametro_toranja = dado[5000:, 0]
peso_laranja = dado[:5000, 1]
peso_toranja = dado[5000:, 1]
```

```python
import matplotlib.pyplot as plt
```

```python
# Uniao do diametro e peso
plt.plot(diametro_laranja, peso_laranja)
plt.plot(diametro_toranja, peso_toranja)
plt.xlabel("Diametro")
plt.ylabel("Peso")
plt.legend(['laranja', 'toranja'])
plt.title("DIAMETRO x PESO")
```

```python
# Laranja
# Calculando a reta linear
Y = peso_laranja
X = diametro_laranja
n = np.size(X)

a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y))/(n*np.sum(X**2) - np.sum(X)**2)
print(a)
b = np.mean(Y) - a*np.mean(X)
print(b)
```

```python
# Toranja
# Calculando a reta linear
Y = peso_toranja
X = diametro_toranja
n = np.size(X)

a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y))/(n*np.sum(X**2) - np.sum(X)**2)
print(a)
b = np.mean(Y) - a*np.mean(X)
print(b)
```

```python
# calculando o coeficiente angular
norma = np.array([])
np.random.seed(84)
coef_angulares = np.random.uniform(low=0.0,high=30.0,size=100)

for i in range(100):
  norma = np.append(norma,np.linalg.norm(Y- (coef_angulares[i]*X+b)))

norma
```

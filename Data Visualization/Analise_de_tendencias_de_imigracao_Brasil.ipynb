```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/alura-cursos/bibliotecas_visualizacao/main/Dados/imigrantes_canada.csv")
df.head(5)
```

```python
df.info()
```

## Vou alterar a ordem dos indies de forma que fique Anos / Imigrantes em duas colunas

```python
# mudando o indice
# INPLACE - vai ser executado direto no dataFrame, sem a necessidade de criar um novo dataFrame
df.set_index('País', inplace=True)
```

```python
# armazena o intervalo de tempo dos anos
# usando o MAP para deixar os valores em STR
anos = list(map(str, range(1980, 2014)))
anos
```

```python
# Vou trabalhar com o Brasil
# LOC pega os rotulos expecifiocs dentro do dataFrame
brasil = df.loc['Brasil', anos]
brasil
```

```python
# criadno um dicionario dando os nomes das colunas
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
#criando o dataFrame
dados_brasil = pd.DataFrame(brasil_dict)
dados_brasil
```

## Criando o grafico

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
```

## Melhorando a visualizaco do grafico

```python
# mudando o tamanho do grafico 
plt.figure(figsize=(8,4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.title("Imigracao do Brasil para o Canada\n 1980 a 2014", fontsize=18)
plt.xlabel('Ano', fontsize=14)
plt.ylabel('Imigrantes', fontsize=14)
# selecionando os anos que devem aparecer
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
# exibindo o grafio
plt.show()
```

## Criando uma figura

```python
# fig - permite criar uma figura
# ax - representa o eixo
fig, ax=plt.subplots(figsize=(8,4))
# lw - expessura
# marker - marcadores (marker='o')
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'], lw=3)
ax.set_title('Imigracao do Brasil para o Canada\n 1980 a 2013', fontsize=18, loc='left')
# add rotulos
ax.set_xlabel('Ano', fontsize=14)
ax.set_ylabel('Numeros de Imigrantes', fontsize=14)
# alterando os valores dos eixos
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
# nao deixa os anos sobrepostos, informando de 5 em 5
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.grid(linestyle='--')
plt.show()
```

## Criando subplots em uma direcao 

```python
# axs - mais de um eixo
fig, axs=plt.subplots(1,2,figsize=(15,5))
axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])
axs[0].set_title('Imigracao do Brasil para o Canada\n 1980 a 2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Numeros de Imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_title('Boxplot da imigracao do Brasil para o Canada\n 1980 a 2013')
axs[1].set_xlabel('Ano')
axs[1].set_ylabel('Numeros de Imigrantes')
axs[1].grid()

plt.show()
```

## Criando subplots em duas direcao

Referente a funcao `fig.suplots_adjust()` que recebe o param `hspace`. Ele controla o espacamento Vertical entre os subplots e o para `wspace`, que controla o epacamento Horizontal

---

**O valor passado para esses parametros e um numero decimal que representa a fracao do tamanho da figura, por exemplo, 0.5 significa que o espaçamento sera de 50% da altura/largura da figura.**

```python
# axs - mais de um eixo
fig, axs = plt.subplots(2,2, figsize=(10,6))
# add espacamento
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Imigracao dos 4 maoires paises da America do Sul para o Canada de 1980 a 2013')

axs[0,0].plot(df.loc['Brasil', anos])
axs[0,0].set_title('Brasil')

axs[0,1].plot(df.loc['Colômbia', anos])
axs[0,1].set_title('Colômbia')

axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Peru', anos])
axs[1,1].set_title('Peru')

# percorrendo todos os eixos
for i in axs.flat:
    i.xaxis.set_major_locator(plt.MultipleLocator(5))
    i.set_xlabel('Ano')
    i.set_ylabel('Numero de imigrantes')
    i.grid()

# transformando em array bidimensional em um unidimencional
ymin= 0
ymax= 7000

for i in axs.ravel():
    i.set_ylim(ymin, ymax)

plt.show()
```

## Alterando cores do grafico

```python
fig, ax = plt.subplots(figsize=(8,4))
# alterando a cor da linha 
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'], lw=3, color='g')
ax.set_title('Imigracao do Brasil para o Canada\n1980 a 2013', fontsize=20, loc='left')
ax.set_xlabel('Ano', fontsize=14)
ax.set_ylabel('Numero de imigrantes', fontsize=14)
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_tick_params(labelsize=12)
ax.xaxis.set_major_locator(plt.MultipleLocator(5))

# removendo as linhas lateral e superior
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```

## Comparando o Brasil com todos os Paises da America do Sul

```python
america_sul = df.query('Região == "América do Sul"')
america_sul
```

```python
# ordenando do maior para o menor
america_sul_ordenado = america_sul.sort_values('Total', ascending=True)
```

```python
cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']

fig, ax = plt.subplots(figsize=(12,5))
# barra de grafico na horizontal
ax.barh(america_sul_ordenado.index, america_sul_ordenado['Total'], color=cores)
ax.set_title('Imigração da America do Sul para o Canadá\n1980 a 2013', fontsize=18, loc='left')
ax.set_xlabel('Numero de Imigrantes', fontsize=14)
ax.set_ylabel('')
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_tick_params(labelsize=12)
plt.show()
```

```python
cores = []

for pais in america_sul_ordenado.index:
    if pais == 'Brasil':
        cores.append('green')
    else:
        cores.append('silver')

fig, ax = plt.subplots(figsize=(12,5))
# barra de grafico na horizontal
ax.barh(america_sul_ordenado.index, america_sul_ordenado['Total'], color=cores)
ax.set_title('América do Sul: Brasil foi o quarto país com mais imigrantes\npara o Canadá no período de 1980 a 2013', fontsize=18, loc='left')
ax.set_xlabel('Numero de Imigrantes', fontsize=14)
ax.set_ylabel('')
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_tick_params(labelsize=12)

# adicionae anotacoes no grafico
for posicao, barra in enumerate(america_sul_ordenado['Total']):
    ax.text(barra+20, posicao, str(barra), color='black', fontsize=10, ha='left', va='center')

# removendo a caia em volta
ax.set_frame_on(False)
# tirando o eixo X
ax.get_xaxis().set_visible(False)
# removendo os tracos
ax.tick_params(axis='both', which='both', length=0)

plt.show()
```

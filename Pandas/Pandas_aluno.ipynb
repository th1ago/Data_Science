Verificando se a base de dados possui dados nulos e, caso tenha, realizar o tratamento desses dados nulod da forma mais coerente.

```python
# soma os dados nulos
df = data
df.isnull().sum()
```

```python
# substituindo todos os dados nulos por Zero
df = df.fillna(0)
df
```

```python
# verificando os dados nulos
df.isnull().sum()
```

Os alunos "Alice" e "Carlos", nao fazem mais parte da turma. Sendo assim, remova-os da base de dados

```python
# selecionando os alunas
aluno_removido = df.query('Nome == "Alice" | Nome == "Carlos"').index
aluno_removido
```

```python
# removendo a linha
df.drop(aluno_removido, axis=0, inplace=True)
df
```
